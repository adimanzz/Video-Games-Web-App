```python
from twitterscraper import query_tweets
import datetime as dt
import numpy as np
import pandas as pd
pd.options.display.max_columns = 500
import matplotlib.pyplot as plt
import seaborn as sns
from attention import Attention
from clean_text import CleanText

import tensorflow as tf
#physical_devices = tf.config.experimental.list_physical_devices('GPU')
#tf.config.experimental.set_memory_growth(physical_devices[0], True)

from keras.models import load_model
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
begin_date = dt.date(2019,5,1)
end_date = dt.date(2019,12,31)
```

```python
limit = 100
lang = 'english'
#user = adimanz
```

```python
tweets = query_tweets('video games', begindate = begin_date, enddate = end_date, limit = limit, lang = lang)
```

```python
len(tweets)
```

```python
df = pd.DataFrame(t.__dict__ for t in tweets)
```

```python
len(df)
```

```python
df = df[['timestamp','text','likes','retweets']]
```

```python
df.head()
```

### Cleaning

```python
clean = CleanText()
```

```python
#clean() removes urls, emoticons and hashtags
df['clean_text'] = clean.clean(df['text']) 
```

```python
df['clean_text'].iloc[0]
```

I have kept '!' and '?' because they help emphasize an sentiment.

```python
#remove punctuations, stopwords, lemmatize and splits the sentences into tokens
df['clean_text'] = df['clean_text'].apply(lambda x: clean.tokenize(x)) 
```

```python
df['clean_text'].iloc[0]
```

All the tweets have been cleaned and tokenized. This will make it easier to vectorize the words and help the model to predict the sentiment of each tweet.

## Word Embeddings

Using pretrained GloVe embeddings to represent the words.

```python
docs = df['clean_text']

#tokenizer
t = Tokenizer()
t.fit_on_texts(docs)
vocab_size = len(t.word_index) + 1

#encode the documents
encoded_docs = t.texts_to_sequences(docs)

#pad docs to max length
padded_docs = pad_sequences(encoded_docs, maxlen = 40, padding = 'post') 
```

```python
# Loading the classifier 
#classifier = load_model('Models/model_sentiment.h5',custom_objects = {'Attention':Attention}) #Negative: 0, Neutral: 1, Postive: 2
classifier = load_model('Models/model_mil_77.h5',custom_objects = {'Attention':Attention})
```

```python
labels_categorical = classifier.predict(padded_docs) # Predicting the Sentiments of the Video game related tweets
```

```python
labels_categorical[:10] #Output of each class by the softmax function
```

```python
np.argmax(labels_categorical[:10], axis = 1) #np.argmax to get labels of the classes Negative: 0, Neutral: 1, Postive: 2
```

```python
df['labels'] = np.argmax(labels_categorical, axis = 1)
```

```python
def label_to_sentiment(label):
    if label == 0:
        return 'Negative'
    else:
        return 'Positive'
```

```python
df['sentiment'] = df['labels'].apply(lambda x: label_to_sentiment(x))
```

```python
from wordcloud import WordCloud
def plot_wordcloud(data):
    words = []
    for sent in data:
        for word in sent:
            words.append(word) 
    words = pd.Series(words).str.cat(sep=' ')
    wordcloud = WordCloud(width=1600, height=800,max_font_size=200).generate(words)
    plt.figure(figsize=(12,10))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()
```

```python
plot_wordcloud(df['clean_text'][df['sentiment'] == 'Positive'])
```

## Topic Modelling using LDA

```python
# Importing Gensim
import gensim
from gensim import corpora

# Creating the term dictionary of our corpus, where every unique term is assigned an index. 
dictionary = corpora.Dictionary(docs)

# Converting list of documents (corpus) into Document Term Matrix using dictionary prepared above.
doc_term_matrix = [dictionary.doc2bow(doc) for doc in docs]
```

```python
# Creating the object for LDA model using gensim library
Lda = gensim.models.ldamodel.LdaModel
```

```python
# Running and Trainign LDA model on the document term matrix.
ldamodel2 = Lda(doc_term_matrix, num_topics=5, id2word = dictionary, passes=10, iterations=10) 
```

```python
for idx, topic in ldamodel2.show_topics(formatted=False, num_words= 10):
    print('Topic: {} \nWords: {}'.format(idx+1, '|'.join([w[0] for w in topic])))
```
